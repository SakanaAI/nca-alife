```python
import numpy as np

def update_rule(grid):
    """
    Apply the custom cellular automata update rule to the input grid.
    
    Parameters:
    grid (numpy.ndarray): 2D array of 0s and 1s representing the initial state of the grid.
    
    Returns:
    numpy.ndarray: Updated grid after applying the rule.
    """
    # Create a copy of the grid to avoid modifying the original
    grid = np.copy(grid)
    grid_padded = np.pad(grid, ((1, 1), (1, 1)), mode='wrap')

    windows = np.lib.stride_tricks.sliding_window_view(grid_padded, (3, 3))
    neighbors = np.sum(windows, axis=(2, 3)) - grid
    a = (grid==0) & ((neighbors==0) | (neighbors==1) | (neighbors==3) | (neighbors==5) | (neighbors==6))
    b = (grid==1) & ((neighbors==0) | (neighbors==1) | (neighbors==2) | (neighbors==3))
    return (a | b).astype(int)
```

```python
import matplotlib.pyplot as plt
from tqdm.auto import tqdm


for seed in range(16):
    np.random.seed(seed)
    grid = (np.random.uniform(0, 1, (100, 100)) < 0.4).astype(int)
    for t in tqdm(range(1000)):
        grid = update_rule(grid)
    plt.imshow(grid)
    plt.show()
```
